```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

```python
total_no_of_schools = school_data_complete["School ID"].nunique()
#print(f"total number of schools:  {total_no_of_schools}")

total_numbe_of_students = len(school_data_complete)
#print(f"total number of students: {total_numbe_of_students}")

#school_data_complete["budget"].value_counts()math_score
#school_data_complete.groupby(by="School ID").sum()
df_unique_budget = pd.DataFrame({"unique_budget":school_data_complete["budget"].unique()})
budget_sum = df_unique_budget.sum()
#print(f"Total School Budget: {budget_sum}")
Average_math_score = school_data_complete["math_score"].mean()
Average_reading_score = school_data_complete["reading_score"].mean()

Overall_Avg_Score = (Average_math_score + Average_reading_score)/2

count_math_gte_70 = school_data_complete.loc[ school_data_complete["math_score"] >= 70, ["math_score"]].count()
per_math_gte_70 = (count_math_gte_70/total_numbe_of_students) * 100

count_reading_gte_70 = school_data_complete.loc[ school_data_complete["reading_score"] >= 70, ["reading_score"]].count()
per_reading_gte_70 = (count_reading_gte_70/total_numbe_of_students) * 100

print(f"{total_no_of_schools} : {total_numbe_of_students} ")
print(f"{Average_math_score} : {Average_reading_score} : {Overall_Avg_Score}")
print(f"{count_math_gte_70} : {per_math_gte_70}")
print(f"{count_reading_gte_70} : {per_reading_gte_70}")
```

## Top Performing Schools (By Passing Rate)

* Sort and display the top five schools in overall passing rate

```python

groupby_counts = school_data_complete.groupby(["school_name"])["school_name"].count()

groupby_read_avg = school_data_complete.groupby(["school_name"])["reading_score"].mean()

groupby_math_avg = school_data_complete.groupby(["school_name"])["math_score"].mean()

groupby_type = school_data_complete.groupby(["school_name"])["type"].max()

groupby_buget = school_data_complete.groupby(["school_name"])["budget"].max()

groupby_per_student = groupby_buget/groupby_counts
#print(f"{groupby_per_student}")

df_math_pass = school_data_complete.loc[ school_data_complete["math_score"] >= 70, :]
groupby_math_pass_count = df_math_pass.groupby(["school_name"])["school_name"].count()
group_math_per = (groupby_math_pass_count/groupby_counts) * 100


df_reading_pass = school_data_complete.loc[ school_data_complete["reading_score"] >= 70, :]
groupby_reading_pass_count = df_reading_pass.groupby(["school_name"])["school_name"].count()
group_reading_per = (groupby_reading_pass_count/groupby_counts) * 100

overall_pass_avg = (group_math_per + group_reading_per)/2

#print(f"{overall_pass_avg}")


df_final = pd.DataFrame({"student_count":groupby_counts, "students_read_avg":groupby_read_avg, "students_math_avg":groupby_math_avg, 
                         "school_budget":groupby_buget, "budget_per_student":groupby_per_student, "School Type":groupby_type,
                        "math_pass_per":group_math_per, "math_reading_per":group_reading_per, "overall_pass_per":overall_pass_avg})
df_best_perform = df_final.sort_values("overall_pass_per", ascending=False)
#df_final.sort_index()
df_best_perform.head(5)
```
